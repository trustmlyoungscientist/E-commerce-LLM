```python
import pandas as pd
import copy
```

```python
all_data = pd.read_json("train-qar.jsonl",lines=True, nrows=200000).sample(frac = 1).reset_index(drop=True)
generation_data_part = all_data[0:100000].reset_index(drop=True)
answerability_data_part = all_data[100000:].reset_index(drop=True)
```

### generation instruction

```python
generation_data = copy.deepcopy(generation_data_part)
template = "Given a question and the related document, and generate the answer to the question based on the information provided in the document.\nQuestion Text:{}\nRelated Review:{}\n\n### Answers: "
generation_data = generation_data[generation_data['is_answerable'] == 1]
generation_data['prompt'] = generation_data.apply(lambda row:template.format(row['questionText'],"\n".join([f"User{idx}:{i}" for idx, i in enumerate(row['review_snippets'][0:5])])),axis =1 )
generation_data['out_put'] = generation_data['answers'].apply(lambda x:x[0]['answerText'])
generation_data[['prompt','out_put']].to_csv("answer_generation_task.csv",index=None)
generation_data[['prompt','out_put']].to_json("answer_generation_task.json",lines = True,orient='records')
generation_data
```

### Generate instruction data for the is_canpred task

```python
answerability_data = copy.deepcopy(answerability_data_part)
template = "Given a question and the related document, predict if the question is answerable based on the information provided in the document. Output only yes or no.\nQuestion Text:{}\nRelated Review:{}\n\n### Answers: "
answerability_data['prompt'] = answerability_data.apply(lambda row:template.format(row['questionText'],"\n".join([f"User{idx}:{i}" for idx, i in enumerate(row['review_snippets'][0:5])])),axis =1 )
answerability_data['out_put'] = answerability_data['is_answerable'].apply(lambda x:x)
answerability_data[['prompt','out_put']].to_csv("answerability_task.csv",index=None)
answerability_data[['prompt','out_put']].to_json("answerability_task.json",lines = True,orient='records')
answerability_data
```
